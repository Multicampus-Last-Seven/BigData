```python
from matplotlib import font_manager, rc
font_path = "data/fontYouandiModernTR.ttf"  
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
from wordcloud import WordCloud            
import matplotlib.pyplot as plt 
```

```python
myfontpath = "data/fontYouandiModernTR.ttf" 
```

```python
wc = WordCloud(                       
    font_path = myfontpath,
    background_color='black',
    colormap = 'Set1',
    width = 1000,
    height = 1000
)
```

```python
f = open("data/news1.txt", 'r', encoding='UTF8')
text = f.read()
```

```python
wc = wc.generate(text)   
wc.to_file('output/fire5_8.png')
f.close()
```

```python
fig = plt.figure()
plt.imshow(wc, interpolation='bilinear')               
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(
    font_path = myfontpath,
    background_color='white',                      
    width = 1000,
    height = 1000
)
wc = wc.generate(text)
fig = plt.figure()
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
